```python
import pandas as pd
```

## 시리즈(Series)

```python
# 시리즈 - 미니퀘스트 1번
# Series를 리스트 [5, 10, 15, 20]을 사용하여 생성하고, Series의 인덱스를 확인하는 코드

series = pd.Series([5, 10, 15, 20])
print(series.index)
```

```python
# 시리즈 - 미니퀘스트 2번
# 딕셔너리를 이용하여 Pandas Series를 생성하고, 인덱스를 활용하여 'b'의 값을 출력하는 코드

series = pd.Series({'a': 100, 'b': 200, 'c': 300})
print(series['b'])
```

```python
# 시리즈 - 미니퀘스트 3번
# Series에 대해 결측값(NaN)을 확인하고, 모든 결측값을 0으로 채운 후 Series의 값을 출력하는 코드

series = pd.Series([1, 2, None, 4, None, 6])

print("--- 결측값 확인 (True가 결측값) ---")
print(series.isna())
print("\n--- 결측값을 0으로 채운 결과 ---")
print(series.fillna(0))
```

## 데이터프레임

```python
# 데이터프레임 - 미니퀘스트 1번
# 데이터프레임에서 열(column)의 이름들을 출력하는 코드

data = {'이름': ['홍길동', '김철수', '박영희'],
        '나이': [25, 30, 28],
        '성별': ['남', '남', '여']}

df = pd.DataFrame(data)
print(df.columns)
```

```python
# 데이터프레임 - 미니퀘스트 2번
# 데이터프레임에서 나이 열(age)을 기준으로 오름차순으로 정렬된 새로운 데이터프레임을 생성하고 출력하는 코드

data = {'이름': ['홍길동', '김철수', '박영희'],
        '나이': [25, 30, 28],
        '성별': ['남', '남', '여']}

df = pd.DataFrame(data)
df_sorted_age = df.sort_values(by='나이')
print(df_sorted_age)
```

```python
# 데이터프레임 - 미니퀘스트 3번
# 각 학생의 총점을 계산하는 새로운 열을 추가한 후, 총점이 250점 이상인 학생들만 포함된 데이터프레임을 생성하는 코드

data = {'이름': ['홍길동', '김철수', '박영희', '이순신'],
        '국어': [85, 90, 88, 92],
        '영어': [78, 85, 89, 87],
        '수학': [92, 88, 84, 90]}
df = pd.DataFrame(data)

df['총점'] = df['국어'] + df['영어'] + df['수학']
df_high_score = df[df['총점'] >= 250]

print(df_high_score)
```

## 필터링(Filtering)

```python
# 필터링 - 미니퀘스트 1번
# 나이가 30 이상인 행만 필터링하는 코드

data = {'이름': ['홍길동', '김철수', '박영희', '이순신', '강감찬'],
        '나이': [25, 30, 35, 40, 45],
        '도시': ['서울', '부산', '서울', '대구', '부산']}

df = pd.DataFrame(data)
print(df[df['나이'] >= 30])
```

```python
# 필터링 - 미니퀘스트 2번
# 도시가 '서울'이거나 점수가 80점 이상인 데이터를 필터링하는 코드

data = {'이름': ['홍길동', '김철수', '박영희', '이순신', '강감찬'],
        '나이': [25, 30, 35, 40, 45],
        '도시': ['서울', '부산', '서울', '대구', '부산'],
        '점수': [85, 90, 75, 95, 80]}

df = pd.DataFrame(data)
print(df[(df['도시'] == '서울') | (df['점수'] >= 80)])
```

```python
# 필터링 - 미니퀘스트 3번
# query()를 사용하여 나이가 35 이상이고 점수가 80점 초과인 데이터를 필터링하는 코드

import pandas as pd

data = {'이름': ['홍길동', '김철수', '박영희', '이순신', '강감찬'],
        '나이': [25, 30, 35, 40, 45],
        '도시': ['서울', '부산', '서울', '대구', '부산'],
        '점수': [85, 90, 75, 95, 80]}

df = pd.DataFrame(data)
print(df.query("나이 >= 35 and 점수 > 80"))
```

## 그룹화(Grouping)

```python
data = {
    '이름': ['홍길동', '김철수', '박영희', '이순신', '강감찬', '신사임당'],
    '부서': ['영업', '영업', '인사', '인사', 'IT', 'IT'],
    '급여': [5000, 5500, 4800, 5100, 6000, 6200]
}

df = pd.DataFrame(data)
```

```python
# 그룹화 - 미니퀘스트 1번
# 데이터프레임에서 groupby()를 사용하여 '부서'별로 급여의 합계를 구하는 코드

department_salary = df.groupby('부서')['급여'].sum()
print(department_salary)
```

```python
# 그룹화 - 미니퀘스트 2번
# 데이터프레임에서 groupby()와 agg()를 사용하여 '부서'별 급여의 합계(sum)와 평균(mean)을 계산하는 코드

grouped = df.groupby('부서')['급여'].agg(['sum', 'mean'])
print(grouped)
```

```python
# 그룹화 - 미니퀘스트 3번
# 부서별로 데이터를 그룹화한 뒤, 각 부서의 평균 급여가 5000 이상인 경우만 필터링하는 코드

filtered = df.groupby('부서').filter(lambda x: x['급여'].mean() > 5000)
print(filtered)
```

##병합(Merging)

```python
# 병합 - 미니퀘스트 1번
# 데이터프레임을 고객ID를 기준으로 내부 조인(inner join)하여 공통된 데이터를 병합하는 코드

df1 = pd.DataFrame({'고객ID': [1, 2, 3], '이름': ['홍길동', '김철수', '이영희']})
df2 = pd.DataFrame({'고객ID': [2, 3, 4], '구매액': [10000, 20000, 30000]})

merged = pd.merge(df1, df2, on='고객ID', how='inner')
print(merged)
```

```python
# 병합 - 미니퀘스트 2번
# 왼쪽 데이터프레임을 기준으로 병합(left join)을 수행하고, 구매액이 없는 경우 NaN을 유지하는 코드

df1 = pd.DataFrame({'고객ID': [1, 2, 3], '이름': ['홍길동', '김철수', '이영희']})
df2 = pd.DataFrame({'고객ID': [2, 3, 4], '구매액': [15000, 25000, 35000]})

merged = pd.merge(df1, df2, how='left')
print(merged)
```

```python
# 병합 - 미니퀘스트 3번
#두 개 이상의 열을 기준으로 병합하고, 중복 열의 이름을 구분하도록 접미사(suffixes)를 지정하는 코드

df1 = pd.DataFrame({
    '고객ID': [1, 2, 3],
    '도시': ['서울', '부산', '대전'],
    '구매액': [10000, 20000, 30000]
})

df2 = pd.DataFrame({
    '고객ID': [1, 2, 3],
    '도시': ['서울', '부산', '광주'],
    '구매액': [15000, 25000, 35000]
})

multiple_merged = pd.merge(df1, df2, on=['고객ID', '도시'], suffixes=('_기존', '_신규'))
print(multiple_merged)
```

##결측치 처리(Missing Data)

```python
import numpy as np

data = {'이름': ['홍길동', '김철수', np.nan, '이영희'],
        '나이': [25, np.nan, 30, 28],
        '성별': ['남', '남', '여', np.nan]}

df = pd.DataFrame(data)
```

```python
# 결측치 처리 - 미니퀘스트 1번
# 데이터프레임에서 결측치를 탐지하고, 열별 결측치 개수를 출력하는 코드

print(df.isnull().sum())
```

```python
# 결측치 처리 - 미니퀘스트 2번
# 데이터프레임에서 결측치가 포함된 행을 삭제하고, 새로운 데이터프레임을 출력하는 코드

new_df = df.dropna()
print(new_df)
```

```python
# 결측치 처리 - 미니퀘스트 3번
# 결측치가 포함된 '나이' 열을 평균값으로 대체하고, 새로운 데이터프레임을 출력하는 코드

age_mean = df['나이'].mean()
new_df = df.copy()
new_df['나이'] = new_df['나이'].fillna(age_mean)
print(new_df)
```

##피벗(Pivot)

```python
# 피벗 - 미니퀘스트 1번
# pivot() 함수를 사용하여 날짜(날짜)를 행 인덱스로, 제품(제품)을 열로, 판매량(판매량)을 값으로 설정하는 코드

data = {
    '날짜': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
    '제품': ['A', 'B', 'A', 'B'],
    '판매량': [100, 200, 150, 250]
}

df = pd.DataFrame(data)
print(df.pivot(index='날짜', columns='제품', values='판매량'))
```

```python
# 피벗 - 미니퀘스트 2번
# pivot_table()을 사용하여 주어진 데이터프레임에서 카테고리(카테고리)를 행 인덱스로, 제품(제품)을 열로 설정하고, 판매량(판매량)의 합계를 출력하는 코드

data = {
    '카테고리': ['전자', '가전', '전자', '가전'],
    '제품': ['A', 'B', 'A', 'B'],
    '판매량': [100, 200, 150, 250]
}

df = pd.DataFrame(data)
```

```python
# 피벗 - 미니퀘스트 3번
# 여러 값을 동시에 피벗하여, pivot()을 사용해 날짜(날짜)를 행으로, 제품(제품)을 열로 설정하고, 판매량(판매량)과 이익(이익)을 동시에 피벗하는 코드

data = {
    '날짜': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
    '제품': ['A', 'B', 'A', 'B'],
    '판매량': [100, 200, 150, 250],
    '이익': [20, 50, 30, 60]
}

df = pd.DataFrame(data)
df_pivot = df.pivot_table(index='날짜', columns='제품', values=['판매량', '이익'])
print(df_pivot)
```

##중복제거

```python
# 중복제거 - 미니퀘스트 1번
# 데이터프레임에서 기본적으로 중복된 행을 제거하는 코드

data = {
    '이름': ['김철수', '이영희', '김철수', '박민수'],
    '나이': [25, 30, 25, 40],
    '성별': ['남', '여', '남', '남']
}

df = pd.DataFrame(data)
df_unique = df.drop_duplicates()
print(df_unique)
```

```python
# 중복제거 - 미니퀘스트 2번
# 데이터프레임에서 특정 열을 기준으로 중복을 제거한 후, 중복 여부를 확인하는 코드

data = {
    '제품': ['노트북', '태블릿', '노트북', '스마트폰'],
    '가격': [1500000, 800000, 1500000, 1000000],
    '카테고리': ['전자기기', '전자기기', '전자기기', '전자기기']
}

df = pd.DataFrame(data)
df_unique_category = df.drop_duplicates(subset=['제품'])
print(df_unique_category)
```

```python
# 중복제거 - 미니퀘스트 3번
# 데이터프레임에서 중복된 모든 행을 삭제한 후, 남은 데이터를 저장하고 다시 불러오는 코드

data = {
    '학생': ['김민수', '박지현', '김민수', '이정훈'],
    '성적': [90, 85, 90, 88],
    '학교': ['A고', 'B고', 'A고', 'C고']
}

df = pd.DataFrame(data)
df_none = df.drop_duplicates()

df_none.to_csv('unique.csv', index=False)
df_loaded = pd.read_csv('unique.csv')

print(df_loaded)
```

##문자열 처리

```python
# 문자열 처리 - 미니퀘스트 1번
# Pandas의 .str 접근자를 사용하여 아래 시리즈의 모든 문자열을 소문자로 변환하는 코드

data = pd.Series(["HELLO", "WORLD", "PYTHON", "PANDAS"])

lowercase_data = data.str.lower()
print(lowercase_data)
```

```python
# 문자열 처리 - 미니퀘스트 2번
# 데이터프레임에서 '이름' 컬럼의 문자열 앞뒤 공백을 제거하고, 특정 문자열 "Doe"가 포함된 행을 필터링하는 코드

df = pd.DataFrame({"이름": [" John Doe ", "Alice ", " Bob", "Charlie Doe "]})

df['이름'] = df['이름'].str.strip()
result = df[df['이름'].str.contains('Doe')]
print(result)
```

```python
# 문자열 처리 - 미니퀘스트 3번
# 데이터프레임에서 '설명' 컬럼의 문자열을 공백을 기준으로 나누고, 각 단어의 첫 글자만 추출하여 새로운 컬럼 '약어'를 생성하는 코드

df = pd.DataFrame({"설명": ["빅데이터 분석", "데이터 과학", "머신 러닝", "딥 러닝"]})

split = df['설명'].str.split()
first_word = split.str[0].str[0]
second_word = split.str[1].str[0]

df['약어'] = first_word + second_word
print(df)
```
